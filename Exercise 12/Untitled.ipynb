```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import KMeans
```

```python
#read data
df = pd.read_csv("edlich-kmeans-A0.csv")
df.head()
```

```python
x = df["V1"]
y = df["V2"]
z = df["V3"]
```

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(x,y,z)
```

```python
kmeans = KMeans(n_clusters=5)
kmeans.fit(df.values)
center = kmeans.cluster_centers_
center
```

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(x,y,z,c=y)
ax.scatter(center[:,0],center[:,1],center[:,2], marker= "*",c="#050505", s=500)
```

```python
# A = Best K is 5
```

```python
# B
count = 0
for v in center:
    count+=1
    print("cluster vector: " + str(count) + " - " + "x: " + str(v[0]) + " y: " + str(v[1]) + " z: " + str(v[2]))
```

```python
# C
for row in df.values:
    pred = kmeans.predict([row])
    print("value: " + str(row[0]) + ", " + str(row[1]) + ", " + str(row[2]) + " = Cluster: " + str(pred))
    
```
